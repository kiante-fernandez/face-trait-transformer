## Upload or point to a face image

```python
from pathlib import Path

try:
    from google.colab import files  # type: ignore
    uploaded = files.upload()
    image_path = Path(next(iter(uploaded)))
except ImportError:
    # Running locally (not on Colab): set this to your own image path.
    image_path = Path('face.jpg')

print('using image:', image_path)
```

## Predict and render the diagnostic panel

```python
from face_trait_transformer import TraitPredictor

predictor = TraitPredictor.from_pretrained()  # defaults to kiante/face-trait-transformer
row, fig = predictor.predict_with_figure(image_path, out_path='diag.png')
print(row.round(1).to_string())
fig
```

## Interpretation

Values are on a **0â€“100** scale and represent the predicted **population-mean rating** that OMI's rater pool would have given this image on each of the 34 perceived-trait dimensions.

**These are perceived traits, not ground truth.** Especially for demographic (`asian`, `black`, `white`, `middle-eastern`, `hispanic`, `islander`, `native`) and socially-constructed dimensions (`gay`, `liberal`, `privileged`, `godly`, `electable`, `looks-like-you`), frame outputs as "perceived X", not "X". See the model card at https://huggingface.co/kiante/face-trait-transformer for the full bias and limitations discussion.
